```python
# Bài 1
# Viết chương trình chuyển tiếng Việt có dấu sang tiếng Việt không có dấu. 
# Ví dụ: Nhập vào là "Nguyễn văn bình" thì kết quả là "Nguyen van binh".
# name = input('Nhập họ tên: ')
name = 'Nguyễn lâm bảo lộc'
sname = ''
def standard_char(x):
    a = ['à', 'á', 'ả', 'ã', 'ạ', 'ă', 'ắ', 'ằ', 'ẳ', 'ẵ', 'ặ', 'â', 'ấ', 'ầ', 'ậ', 'ẫ']
    e = ['è', 'é', 'ẻ', 'ẽ', 'ẹ', 'ê', 'ế', 'ề', 'ể', 'ễ', 'ệ']
    i = ['ì', 'í', 'ỉ', 'ĩ', 'ị']
    y = ['ỳ', 'ý', 'ỷ', 'ỹ', 'ỵ']
    o = ['ò', 'ó', 'ỏ', 'õ', 'ọ', 'ô', 'ố', 'ồ', 'ổ', 'ỗ', 'ộ', 'ơ', 'ớ', 'ờ', 'ở', 'ỡ', 'ợ']
    u = ['ù', 'ú', 'ủ', 'ũ', 'ụ']
    sx = ''
    if x in a: sx = 'a'
    elif x in e: sx = 'e'
    elif x in i: sx = 'i'
    elif x in y: sx = 'y'
    elif x in o: sx = 'o'
    elif x in u: sx = 'u'
    else: sx = x
    return sx
name = name.lower()
for i in name: sname += standard_char(i)
print(sname)
```

```python
# Bài 2: Trong các công ty sản xuất, 
# thời gian trong tháng được chia làm các tuần. 
# Một tuần đầy đủ (gọi là tuần chẵn) được tính từ thứ 2 đến chủ nhật (có đầy đủ các ngày), 
# còn tuần lẻ là các tuần không đủ 7 ngày. 
# Nhập vào tháng và năm, tính số tuần (bao gồm chẵn và lẻ) trong tháng đó.
import calendar
import datetime
import turtle

# thang = int(input('Nhập số tháng: '))
# nam = int(input('Nhập số năm: '))
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
day_headings = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
thang = 8
nam = 2022
c = 40

def getWeeksInMonth(thang,nam):
    songay = calendar.monthrange(nam, thang)[1]
    days = ''
    for i in range(1, songay+1, 1):
        wd = datetime.datetime(nam, thang, i).weekday()
        if wd != 6:
            days += str(wd)
        else:
            days += str(wd) + ' '
    return days

def setupTurtle():
    t.shape('turtle')
    t.pencolor('gray')
    t.color('black')
    t.pensize(1)
    t.penup()

def calendarDraw(thang,nam):
    t.color('black')
    t.goto(-5.85*c, 3*c)
    for i in range(0,7,1):
        if i < 6:
            t.write(day_headings[i], move=True, align='left', font=("Arial", 12, "bold"))
        else:
            t.color('red')
            t.write(day_headings[i], move=True, align='left', font=("Arial", 12, "bold"))
        t.forward(c*1.15)
    t.color('gray')
    t.goto(-6*c, 2.75*c)
    t.pendown()
    t.goto(6*c, 2.75*c)
    t.penup()
    t.goto(-6*c, 2*c)

    days = getWeeksInMonth(thang,nam)
    weeks = days.strip().split()
    t.color('black')
    stt = 1
    for idw, w in enumerate(weeks):
        t.goto(-5.7*c, (2-idw)*c)
        start = int(w[0])
        stop = int(w[-1])
        for i in range(0,7,1):
            if i >= start and i <= stop:
                t.color('black') if i != 6 else t.color('red')
                t.write(str(stt).zfill(2), move=True, align='left', font=("Arial", 12, "bold"))
                stt += 1
            else:
                t.color('white')
                t.write('00', move=True, align='left', font=("Arial", 12, "bold"))
            t.forward(c*1.45)


def calendarTitle():
    t.goto(6*c, 6*c)
    t.pendown()
    t.write(months[thang-1], move=False, align='right', font=("Arial", 45, "bold"))
    t.goto(6*c, 5*c)
    t.color('red')
    t.write(nam, move=False, align='right', font=("Arial", 30, "normal"))
    t.penup()
    

if thang <= 0 or thang > 13 or nam <= 0:
    print(f'Lỗi nhập dữ liệu {thang} - {nam}')
else:
    
    # vẽ cờ
    t = turtle.Turtle()
    setupTurtle()
    calendarTitle()
    calendarDraw(thang,nam)
    turtle.done()

```
